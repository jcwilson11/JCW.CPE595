## Question 3
Generate an array with 100 rows and 2 cols at random. Plot it by Matplotlib. You should consider the plotting setting in terms of titles, legends, names of the x and y axe, etc.

**Firstly, please set your student id as the seed number. Like np.random.seed(20016329).**

```python
np.random.seed(20006584) # Please use your student id as the seed number
data = np.random.rand(100, 2)
plt.scatter(data[:, 0], data[:, 1], color='blue', label="Random Points")
# Plot settings
plt.title("Scatter Plot of Randomly Generated Data", fontsize=14)
plt.xlabel("feature 1", fontsize=12)
plt.ylabel("feature 2", fontsize=12)
plt.legend()

# Show plot
plt.show()
```

## Question 4
For the previous generated data, consider each row of the array as an individual data point, with the two columns representing its distinct features. Identify the data points with the smallest and largest sum of features. Mark them in red and blue in the figure respectively.

Hint: You can compute the sum of each row and store it in an array named "sums". Then find the minimum or maximum index of this array.

```python
sums = np.sum(data, axis=1)
min_index = np.argmin(sums)
max_index = np.argmax(sums)

plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], color='gray', label="Random Points")
plt.scatter(data[min_index, 0], data[min_index, 1], color='red', s=100, label="Min Sum Point")
plt.scatter(data[max_index, 0], data[max_index, 1], color='blue', s=100, label="Max Sum Point")
plt.title("Scatter Plot of Randomly Generated Data", fontsize=14)
plt.xlabel("feature 1", fontsize=12)
plt.ylabel("feature 2", fontsize=12)
plt.legend()
plt.show()
```

## Question 5
Add two lines representing the mean of the Feature 1 and the 25% quantile of the Feature 2 seperately. Add the legend of the two lines.

```python
sums = np.sum(data, axis=1)
min_index = np.argmin(sums)
max_index = np.argmax(sums)

plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], color='gray', label="Random Points")
plt.scatter(data[min_index, 0], data[min_index, 1], color='red', s=100, label="Min Sum Point")
plt.scatter(data[max_index, 0], data[max_index, 1], color='blue', s=100, label="Max Sum Point")

mean_feature1 = np.mean(data[:, 0])
quantile_25_feature2 = np.quantile(data[:, 1], 0.25)
plt.axvline(x=mean_feature1, color='green', linestyle='--', label="Mean of Feature 1")
plt.axhline(y=quantile_25_feature2, color='purple', linestyle='--', label="25% Quantile of Feature 2")

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3, fontsize=10, frameon=True)
plt.show()
```
